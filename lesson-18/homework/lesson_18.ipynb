```python
import pandas as pd

tack = pd.read_csv('tackoverflow_qa.csv')
tack
```

```python
questions_before_2014 = tack[tack['creationdate'].year < 2014]
```

```python
questions_score_gt_50 = tack[tack['score'] > 50]
print(len(questions_score_gt_50))
print(questions_score_gt_50[['id', 'score', 'title']].head())
```

```python
questions_score_50_to_100 = tack[(tack['score'] >= 50) & (tack['score'] < 100)]
print(len(questions_score_50_to_100))
print(questions_score_50_to_100[['id', 'score', 'title']].head())
```

```python
questions_answered_by_scott = tack[tack['ans_name'] == 'Scott Boston']
print('Scott Boston' in tack['ans_name'].values)
```

```python
top_answerers = ['Wes McKinney', 'unutbu', 'EdChum', 'DSM', 'Alex Riley']
questions_answered_by_top5 = tack[tack['ans_name'].isin(top_answerers)]
print(len(questions_answered_by_top5))
print(questions_answered_by_top5[['id', 'ans_name', 'title']].head())
```

```python
questions_2014_unutbu_low_score = tack[
    (tack['creationdate'] >= '2014-03-01') &
    (tack['creationdate'] <= '2014-10-31') &
    (tack['ans_name'] == 'unutbu') &
    (tack['score'] < 5)
]
print(questions_2014_unutbu_low_score[['id', 'creationdate', 'score', 'title']])
```
